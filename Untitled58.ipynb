```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/content/Book2.csv")
data
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
data.groupby('penyebab').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
jumlah_kategori = data['penyebab'].value_counts()
jumlah_kategori
```

```python
data_penyebab = data.groupby('penyebab').size()
plt.pie(data_penyebab, labels=data_penyebab.index, autopct='%1.1f%%', colors=sns.color_palette('Dark2'), startangle=90)
plt.axis('equal')  # Membuat lingkaran agar pie chart tampak bundar
plt.title('Distribusi Penyebab Bencana')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
data['tahun'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.title('Jumlah Bencana per Tahun')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Bencana')
plt.show()
```

```python
plt.figure(figsize=(14, 6))
data['kecamatan'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Jumlah Bencana per Kecamatan')
plt.xlabel('Kecamatan')
plt.ylabel('Jumlah Bencana')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
data['bencana'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['lightcoral', 'lightgreen', 'lightskyblue'])
plt.title('Persentase Jenis Bencana')
plt.ylabel('')
plt.show()
```

```python
data['jumlah_korban'] = data['meninggal'] + data['hilang'] + data['terluka']

# Membuat dataframe untuk heatmap
heatmap_data = data.groupby(['tahun'])[['meninggal', 'hilang', 'terluka', 'jumlah_korban']].sum().reset_index()

# Visualisasi: Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data[['meninggal', 'hilang', 'terluka']].transpose(), cmap='Blues', annot=True, fmt='g')
plt.title('Heatmap Jumlah Meninggal, Hilang, dan Terluka per Tahun')
plt.xlabel('Tahun')
plt.ylabel('Jumlah')
plt.show()
```

```python
# Visualisasi: Bar Chart
plt.figure(figsize=(12, 8))

# Bar untuk jumlah meninggal
plt.bar(heatmap_data['tahun'], heatmap_data['meninggal'], color='red', label='Meninggal')

# Bar untuk jumlah hilang
plt.bar(heatmap_data['tahun'], heatmap_data['hilang'], bottom=heatmap_data['meninggal'], color='orange', label='Hilang')

# Bar untuk jumlah terluka
plt.bar(heatmap_data['tahun'], heatmap_data['terluka'], bottom=heatmap_data['meninggal'] + heatmap_data['hilang'], color='yellow', label='Terluka')

# Bar untuk jumlah korban
plt.bar(heatmap_data['tahun'], heatmap_data['jumlah_korban'], color='blue', alpha=0.5, label='Jumlah Korban')

plt.title('Jumlah Meninggal, Hilang, Terluka, dan Jumlah Korban per Tahun')
plt.xlabel('Tahun')
plt.ylabel('Jumlah')
plt.legend()
plt.show()
```

```python
# Membuat dataframe untuk jumlah rumah rusak dan terendam per jenis bencana
house_damage_data = data.groupby(['bencana'])[['rumah_rusak', 'rumah_terendam']].sum().reset_index()

# Visualisasi: Bar Chart
plt.figure(figsize=(10, 6))
house_damage_data.plot(x='bencana', y=['rumah_rusak', 'rumah_terendam'], kind='bar', stacked=True, color=['lightcoral', 'lightblue'])
plt.title('Jumlah Rumah Rusak dan Rumah Terendam per Jenis Bencana')
plt.xlabel('Jenis Bencana')
plt.ylabel('Jumlah')
plt.xticks(rotation=45, ha='right')
plt.show()
```
